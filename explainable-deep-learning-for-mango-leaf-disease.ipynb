```python
# ====== Libraries needed for AlexNet (TensorFlow + Kaggle) ======

# core
import os, glob
import numpy as np

# tensorflow / keras
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# splitting + evaluation
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

# reproducibility
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)


# =========================
# 1) PATHS / PARAMS
# =========================
DATA_DIR = r"/kaggle/input/mangoleafdisease-classes/MangoLeafBD Dataset/MangoLeafBD Dataset"
# change the folder name to whatever shows in /kaggle/input

IMG_SIZE = (224, 224)     # keep 224 for AlexNet/ViT/DaViT fairness
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

# =========================
# 2) READ FILE PATHS + LABELS
# =========================
class_names = sorted([
    d for d in os.listdir(DATA_DIR)
    if os.path.isdir(os.path.join(DATA_DIR, d))
])
num_classes = len(class_names)
print("Classes:", class_names)

label_to_index = {c: i for i, c in enumerate(class_names)}

filepaths, labels = [], []
exts = (".jpg", ".jpeg", ".png", ".bmp", ".webp")

for c in class_names:
    cdir = os.path.join(DATA_DIR, c)
    for f in os.listdir(cdir):
        if f.lower().endswith(exts):
            filepaths.append(os.path.join(cdir, f))
            labels.append(label_to_index[c])

filepaths = np.array(filepaths)
labels = np.array(labels)

print("Total images:", len(filepaths))

# =========================
# 3) STRATIFIED SPLIT
# =========================
from sklearn.model_selection import train_test_split

train_files, temp_files, train_labels, temp_labels = train_test_split(
    filepaths, labels, test_size=0.30, stratify=labels, random_state=SEED
)
val_files, test_files, val_labels, test_labels = train_test_split(
    temp_files, temp_labels, test_size=0.50, stratify=temp_labels, random_state=SEED
)

print("Train:", len(train_files))
print("Val:", len(val_files))
print("Test:", len(test_files))


# =========================
# 4) DATASET LOADER  (ONLY CHANGE: DenseNet201 preprocessing)
# =========================
def load_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img.set_shape([None, None, 3])
    img = tf.image.resize(img, IMG_SIZE)
    img = tf.cast(img, tf.float32)

    # DenseNet201 preprocessing  ✅ (replaces EfficientNetV2)
    img = keras.applications.densenet.preprocess_input(img)

    label = tf.one_hot(label, num_classes)
    return img, label

def make_ds(files, labs, training=False):
    ds = tf.data.Dataset.from_tensor_slices((files, labs))
    if training:
        ds = ds.shuffle(len(files), seed=SEED)
    ds = ds.map(load_image, num_parallel_calls=AUTOTUNE)
    ds = ds.batch(BATCH_SIZE, drop_remainder=training).prefetch(AUTOTUNE)
    return ds

train_ds = make_ds(train_files, train_labels, training=True)
val_ds   = make_ds(val_files, val_labels, training=False)
test_ds  = make_ds(test_files, test_labels, training=False)


# =========================
# 5) AUGMENTATION (effective, not too heavy)
# =========================
augment = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.10),
    layers.RandomZoom(0.12),
    layers.RandomContrast(0.08),
    layers.RandomTranslation(0.05, 0.05),
], name="augmentation")


# =========================
# 7) BUILD DenseNet201 MODEL (improved head for best acc)
# =========================
base = keras.applications.DenseNet201(
    include_top=False,
    weights="imagenet",
    input_shape=(*IMG_SIZE, 3)
)
base.trainable = False

inputs = keras.Input(shape=(*IMG_SIZE, 3))
x = augment(inputs, training=True)
x = base(x, training=False)

# ---- Improved classification head ----
x = layers.BatchNormalization()(x)

# add a light channel-attention (SE-style) block
se = layers.GlobalAveragePooling2D()(x)
se = layers.Dense(se.shape[-1] // 8, activation="relu")(se)
se = layers.Dense(x.shape[-1], activation="sigmoid")(se)
se = layers.Reshape((1, 1, x.shape[-1]))(se)
x = layers.Multiply()([x, se])

# pooling + stronger dense stack
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.40)(x)

x = layers.Dense(512, activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.30)(x)

x = layers.Dense(256, activation="relu")(x)
x = layers.Dropout(0.25)(x)

outputs = layers.Dense(num_classes, activation="softmax")(x)

model = keras.Model(inputs, outputs, name="DenseNet201")
model.summary()



# =========================
# 8) CLASS WEIGHTS (helps if imbalance)
# =========================
from sklearn.utils.class_weight import compute_class_weight
cw = compute_class_weight("balanced", classes=np.arange(num_classes), y=train_labels)
class_weights = {i: w for i, w in enumerate(cw)}
print("Class weights:", class_weights)


# =========================
# 9) STAGE 1 TRAIN (freeze backbone)
# =========================
loss_stage1 = keras.losses.CategoricalCrossentropy(label_smoothing=0.02)

model.compile(
    optimizer=keras.optimizers.AdamW(learning_rate=3e-4, weight_decay=1e-4),
    loss=loss_stage1,
    metrics=["accuracy"]
)

cbs1 = [
    keras.callbacks.ModelCheckpoint("best_stage1.keras", save_best_only=True,
                                    monitor="val_accuracy", mode="max"),
    keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5,
                                      patience=2, verbose=1),
    keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=5,
                                  restore_best_weights=True),
]

history1 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=12,
    callbacks=cbs1,
    class_weight=class_weights
)

# =========================
# 10) STAGE 2 FINE-TUNE (unfreeze more)
# (KEEP SAME STYLE, now works because base exists)
# =========================
base.trainable = True
for layer in base.layers[:-60]:
    layer.trainable = False

model.compile(
    optimizer=keras.optimizers.AdamW(learning_rate=5e-5, weight_decay=5e-6),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

cbs2 = [
    keras.callbacks.ModelCheckpoint("best_finetune.keras", save_best_only=True,
                                    monitor="val_accuracy", mode="max"),
    keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5,
                                      patience=2, verbose=1),
    keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=7,
                                  restore_best_weights=True),
]

history2 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=25,
    callbacks=cbs2,
    class_weight=class_weights
)


# =========================
# 11) NORMAL TEST ACC
# =========================
model = keras.models.load_model("best_finetune.keras", compile=False)
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

test_loss, test_acc = model.evaluate(test_ds, verbose=1)
print("TEST ACC (normal):", test_acc)

# =========================
# 12) TEST-TIME AUGMENTATION (TTA) ACC  [FIXED for TF 2.18 / Keras 3]
#    (no keras layers inside ds.map)
# =========================
tta_aug = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.06),
    layers.RandomZoom(0.06),
], name="tta_aug")

def tta_predict(model, ds, runs=8):
    preds_sum = None
    for r in range(runs):
        preds_run = []
        for x_batch, _ in ds:
            x_aug = tta_aug(x_batch, training=True)   # augment in eager mode
            preds = model.predict(x_aug, verbose=0)
            preds_run.append(preds)
        preds_run = np.concatenate(preds_run, axis=0)

        if preds_sum is None:
            preds_sum = preds_run
        else:
            preds_sum += preds_run

    return preds_sum / runs

tta_preds = tta_predict(model, test_ds, runs=8)
tta_pred_labels = np.argmax(tta_preds, axis=1)

# true labels (for one-hot categorical labels)
true_labels = []
for _, y in test_ds:
    true_labels.extend(np.argmax(y.numpy(), axis=1))
true_labels = np.array(true_labels)

tta_acc = np.mean(true_labels == tta_pred_labels)
print("TEST ACC (TTA):", tta_acc)


import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report

# =========================
# 1) ACCURACY + LOSS GRAPHS
# =========================
def plot_history(histories):
    acc, val_acc, loss, val_loss = [], [], [], []

    for h in histories:
        acc      += h.history.get("accuracy", [])
        val_acc  += h.history.get("val_accuracy", [])
        loss     += h.history.get("loss", [])
        val_loss += h.history.get("val_loss", [])

    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 8))
    plt.rcParams.update({"font.size": 16})
    plt.plot(epochs, acc, label="Train Accuracy", linewidth=2)
    plt.plot(epochs, val_acc, label="Val Accuracy", linewidth=2)
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.title("Training vs Validation Accuracy")
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(12, 8))
    plt.rcParams.update({"font.size": 16})
    plt.plot(epochs, loss, label="Train Loss", linewidth=2)
    plt.plot(epochs, val_loss, label="Val Loss", linewidth=2)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title("Training vs Validation Loss")
    plt.legend()
    plt.grid(True)
    plt.show()


# If you trained in 2 stages:
plot_history([history1, history2])

# If you trained only once, use:
# plot_history([history])

# =========================
# 2) CONFUSION MATRIX + REPORT
#    (works for BOTH one-hot and sparse labels)
# =========================
y_true = []
y_pred = []

for x_batch, y_batch in test_ds:
    preds = model.predict(x_batch, verbose=0)

    # handle label format automatically
    if len(y_batch.shape) > 1 and y_batch.shape[-1] > 1:
        # one-hot labels
        y_true.extend(np.argmax(y_batch.numpy(), axis=1))
    else:
        # sparse integer labels
        y_true.extend(y_batch.numpy().astype(int))

    y_pred.extend(np.argmax(preds, axis=1))

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(12, 8))
plt.rcParams.update({"font.size": 16})
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix")
plt.colorbar()

tick_marks = np.arange(num_classes)
plt.xticks(tick_marks, class_names, rotation=45, ha="right")
plt.yticks(tick_marks, class_names)

# write numbers in cells
thresh = cm.max() / 2.0
for i in range(num_classes):
    for j in range(num_classes):
        plt.text(
            j, i, cm[i, j],
            ha="center", va="center",
            color="white" if cm[i, j] > thresh else "black"
        )

plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.tight_layout()
plt.show()


import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# =========================
# ROC EVALUATION (Multiclass OvR) - ROBUST VERSION
# Works for BOTH one-hot and sparse labels
# =========================

y_true_list = []
y_score_list = []

for x_batch, y_batch in test_ds:
    # predictions
    y_score_list.append(model.predict(x_batch, verbose=0))

    # handle label format
    yb = y_batch.numpy()
    if len(yb.shape) > 1 and yb.shape[-1] > 1:
        # one-hot
        y_true_list.append(yb)
    else:
        # sparse -> binarize
        y_true_list.append(
            label_binarize(yb.astype(int).reshape(-1), classes=range(num_classes))
        )

y_true  = np.concatenate(y_true_list, axis=0)    # (N, C)
y_score = np.concatenate(y_score_list, axis=0)   # (N, C)

# =========================
# Compute ROC for each class
# =========================
fpr, tpr, roc_auc = {}, {}, {}

for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Micro-average ROC
fpr["micro"], tpr["micro"], _ = roc_curve(y_true.ravel(), y_score.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Macro-average ROC
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(num_classes)]))
mean_tpr = np.zeros_like(all_fpr)

for i in range(num_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

mean_tpr /= num_classes
fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

# =========================
# PLOT ROC
# =========================
plt.figure(figsize=(12, 8))
plt.rcParams.update({"font.size": 19})

for i, cls in enumerate(class_names):
    plt.plot(
        fpr[i], tpr[i],
        linewidth=2,
        label=f"{cls} (AUC = {roc_auc[i]:.3f})"
    )

plt.plot(
    fpr["micro"], tpr["micro"],
    linestyle="--", linewidth=3,
    label=f"Micro-average (AUC = {roc_auc['micro']:.3f})"
)
plt.plot(
    fpr["macro"], tpr["macro"],
    linestyle="--", linewidth=3,
    label=f"Macro-average (AUC = {roc_auc['macro']:.3f})"
)

plt.plot([0, 1], [0, 1], linestyle=":", linewidth=2)

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Multiclass ROC Curve (One-vs-Rest)")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
# ====== Libraries needed for LeafNet (TensorFlow + Kaggle) ======

# core
import os, glob
import numpy as np

# tensorflow / keras
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# splitting + evaluation
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.utils.class_weight import compute_class_weight

# reproducibility
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)

# =========================
# 1) PATHS / PARAMS
# =========================
DATA_DIR = r"/kaggle/input/mangoleafdisease-classes/MangoLeafBD Dataset/MangoLeafBD Dataset"

IMG_SIZE   = (224, 224)
BATCH_SIZE = 32
AUTOTUNE   = tf.data.AUTOTUNE


# =========================
# 2) READ FILE PATHS + LABELS
# =========================
class_names = sorted([
    d for d in os.listdir(DATA_DIR)
    if os.path.isdir(os.path.join(DATA_DIR, d))
])
num_classes = len(class_names)
print("Classes:", class_names)

label_to_index = {c: i for i, c in enumerate(class_names)}

filepaths, labels = [], []
exts = (".jpg", ".jpeg", ".png", ".bmp", ".webp")

for c in class_names:
    cdir = os.path.join(DATA_DIR, c)
    for f in os.listdir(cdir):
        if f.lower().endswith(exts):
            filepaths.append(os.path.join(cdir, f))
            labels.append(label_to_index[c])

filepaths = np.array(filepaths)
labels = np.array(labels)

print("Total images:", len(filepaths))


# =========================
# 3) STRATIFIED SPLIT (70-15-15)
# =========================
train_files, temp_files, train_labels, temp_labels = train_test_split(
    filepaths, labels, test_size=0.30, stratify=labels, random_state=SEED
)
val_files, test_files, val_labels, test_labels = train_test_split(
    temp_files, temp_labels, test_size=0.50, stratify=temp_labels, random_state=SEED
)

print("Train:", len(train_files))
print("Val:", len(val_files))
print("Test:", len(test_files))


# =========================
# 4) DATASET LOADER (LeafNet preprocessing)
# =========================
def load_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img.set_shape([None, None, 3])
    img = tf.image.resize(img, IMG_SIZE)
    img = tf.cast(img, tf.float32)

    # LeafNet preprocessing ✅ (simple rescale)
    img = img / 255.0

    label = tf.one_hot(label, num_classes)
    return img, label

def make_ds(files, labs, training=False):
    ds = tf.data.Dataset.from_tensor_slices((files, labs))
    if training:
        ds = ds.shuffle(len(files), seed=SEED)
    ds = ds.map(load_image, num_parallel_calls=AUTOTUNE)
    ds = ds.batch(BATCH_SIZE, drop_remainder=training).prefetch(AUTOTUNE)
    return ds

train_ds = make_ds(train_files, train_labels, training=True)
val_ds   = make_ds(val_files, val_labels, training=False)
test_ds  = make_ds(test_files, test_labels, training=False)


# =========================
# 5) AUGMENTATION (effective, not too heavy)
# =========================
augment = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.10),
    layers.RandomZoom(0.12),
    layers.RandomContrast(0.08),
    layers.RandomTranslation(0.05, 0.05),
], name="augmentation")


# =========================
# 7) BUILD LeafNet MODEL
# =========================
inputs = keras.Input(shape=(*IMG_SIZE, 3))

x = augment(inputs, training=True)

# ---- LeafNet backbone ----
# Block 1
x = layers.Conv2D(32, 3, padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.Conv2D(32, 3, padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D(2)(x)
x = layers.Dropout(0.20)(x)

# Block 2
x = layers.Conv2D(64, 3, padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.Conv2D(64, 3, padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D(2)(x)
x = layers.Dropout(0.25)(x)

# Block 3
x = layers.Conv2D(128, 3, padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.Conv2D(128, 3, padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D(2)(x)
x = layers.Dropout(0.30)(x)

# Block 4
x = layers.Conv2D(256, 3, padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.Conv2D(256, 3, padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D(2)(x)
x = layers.Dropout(0.35)(x)

# ---- Classification head ----
x = layers.GlobalAveragePooling2D()(x)

x = layers.Dense(
    512, activation="relu",
    kernel_regularizer=keras.regularizers.l2(1e-4)
)(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.40)(x)

x = layers.Dense(
    256, activation="relu",
    kernel_regularizer=keras.regularizers.l2(1e-4)
)(x)
x = layers.Dropout(0.30)(x)

outputs = layers.Dense(num_classes, activation="softmax")(x)

model = keras.Model(inputs, outputs, name="LeafNet")
model.summary()


# =========================
# 8) CLASS WEIGHTS (helps if imbalance)
# =========================
cw = compute_class_weight("balanced", classes=np.arange(num_classes), y=train_labels)
class_weights = {i: w for i, w in enumerate(cw)}
print("Class weights:", class_weights)


# =========================
# 9) STAGE 1 TRAIN
# =========================
loss_stage1 = keras.losses.CategoricalCrossentropy(label_smoothing=0.02)

model.compile(
    optimizer=keras.optimizers.AdamW(learning_rate=3e-4, weight_decay=1e-4),
    loss=loss_stage1,
    metrics=["accuracy"]
)

cbs1 = [
    keras.callbacks.ModelCheckpoint("best_stage1.keras", save_best_only=True,
                                    monitor="val_accuracy", mode="max"),
    keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5,
                                      patience=2, verbose=1),
    keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=5,
                                  restore_best_weights=True),
]

history1 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=12,
    callbacks=cbs1,
    class_weight=class_weights
)


# =========================
# 10) STAGE 2 FINE-TUNE
# (LeafNet has no ImageNet backbone, so just continue with lower LR)
# =========================
model.compile(
    optimizer=keras.optimizers.AdamW(learning_rate=5e-5, weight_decay=5e-6),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

cbs2 = [
    keras.callbacks.ModelCheckpoint("best_finetune.keras", save_best_only=True,
                                    monitor="val_accuracy", mode="max"),
    keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5,
                                      patience=2, verbose=1),
    keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=7,
                                  restore_best_weights=True),
]

history2 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=25,
    callbacks=cbs2,
    class_weight=class_weights
)


# =========================
# 11) NORMAL TEST ACC
# =========================
model = keras.models.load_model("best_finetune.keras", compile=False)
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

test_loss, test_acc = model.evaluate(test_ds, verbose=1)
print("TEST ACC (normal):", test_acc)


# =========================
# 12) TEST-TIME AUGMENTATION (TTA) ACC
# =========================
tta_aug = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.06),
    layers.RandomZoom(0.06),
], name="tta_aug")

def tta_predict(model, ds, runs=8):
    preds_sum = None
    for _ in range(runs):
        preds_run = []
        for x_batch, _ in ds:
            x_aug = tta_aug(x_batch, training=True)
            preds = model.predict(x_aug, verbose=0)
            preds_run.append(preds)
        preds_run = np.concatenate(preds_run, axis=0)
        preds_sum = preds_run if preds_sum is None else preds_sum + preds_run
    return preds_sum / runs

tta_preds = tta_predict(model, test_ds, runs=8)
tta_pred_labels = np.argmax(tta_preds, axis=1)

true_labels = []
for _, y in test_ds:
    true_labels.extend(np.argmax(y.numpy(), axis=1))
true_labels = np.array(true_labels)

tta_acc = np.mean(true_labels == tta_pred_labels)
print("TEST ACC (TTA):", tta_acc)


# =========================
# 13) ACCURACY + LOSS GRAPHS
# =========================
import matplotlib.pyplot as plt

def plot_history(histories):
    acc, val_acc, loss, val_loss = [], [], [], []
    for h in histories:
        acc      += h.history.get("accuracy", [])
        val_acc  += h.history.get("val_accuracy", [])
        loss     += h.history.get("loss", [])
        val_loss += h.history.get("val_loss", [])

    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 8))
    plt.rcParams.update({"font.size": 16})
    plt.plot(epochs, acc, label="Train Accuracy", linewidth=2)
    plt.plot(epochs, val_acc, label="Val Accuracy", linewidth=2)
    plt.xlabel("Epochs"); plt.ylabel("Accuracy")
    plt.title("Training vs Validation Accuracy")
    plt.legend(); plt.grid(True); plt.show()

    plt.figure(figsize=(12, 8))
    plt.rcParams.update({"font.size": 16})
    plt.plot(epochs, loss, label="Train Loss", linewidth=2)
    plt.plot(epochs, val_loss, label="Val Loss", linewidth=2)
    plt.xlabel("Epochs"); plt.ylabel("Loss")
    plt.title("Training vs Validation Loss")
    plt.legend(); plt.grid(True); plt.show()

plot_history([history1, history2])


# =========================
# 14) CONFUSION MATRIX + REPORT
# =========================
y_true, y_pred = [], []

for x_batch, y_batch in test_ds:
    preds = model.predict(x_batch, verbose=0)
    y_true.extend(np.argmax(y_batch.numpy(), axis=1))
    y_pred.extend(np.argmax(preds, axis=1))

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(12, 8))
plt.rcParams.update({"font.size": 16})
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix")
plt.colorbar()

tick_marks = np.arange(num_classes)
plt.xticks(tick_marks, class_names, rotation=45, ha="right")
plt.yticks(tick_marks, class_names)

thresh = cm.max() / 2.0
for i in range(num_classes):
    for j in range(num_classes):
        plt.text(j, i, cm[i, j],
                 ha="center", va="center",
                 color="white" if cm[i, j] > thresh else "black")

plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.tight_layout()
plt.show()


# =========================
# 15) ROC EVALUATION (Multiclass OvR)
# =========================
y_true_list, y_score_list = [], []

for x_batch, y_batch in test_ds:
    y_true_list.append(y_batch.numpy())
    y_score_list.append(model.predict(x_batch, verbose=0))

y_true  = np.concatenate(y_true_list, axis=0)
y_score = np.concatenate(y_score_list, axis=0)

fpr, tpr, roc_auc = {}, {}, {}
for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

fpr["micro"], tpr["micro"], _ = roc_curve(y_true.ravel(), y_score.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

all_fpr = np.unique(np.concatenate([fpr[i] for i in range(num_classes)]))
mean_tpr = np.zeros_like(all_fpr)
for i in range(num_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
mean_tpr /= num_classes

fpr["macro"], tpr["macro"] = all_fpr, mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

plt.figure(figsize=(12, 8))
plt.rcParams.update({"font.size": 19})

for i, cls in enumerate(class_names):
    plt.plot(fpr[i], tpr[i], linewidth=2,
             label=f"{cls} (AUC = {roc_auc[i]:.3f})")

plt.plot(fpr["micro"], tpr["micro"], linestyle="--", linewidth=3,
         label=f"Micro-average (AUC = {roc_auc['micro']:.3f})")
plt.plot(fpr["macro"], tpr["macro"], linestyle="--", linewidth=3,
         label=f"Macro-average (AUC = {roc_auc['macro']:.3f})")

plt.plot([0, 1], [0, 1], linestyle=":", linewidth=2)

plt.xlim([0.0, 1.0]); plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Multiclass ROC Curve (One-vs-Rest)")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
# ====== Libraries needed for AlexNet (TensorFlow + Kaggle) ======

# core
import os, glob
import numpy as np

# tensorflow / keras
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# splitting + evaluation
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

# reproducibility
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)


# =========================
# 1) PATHS / PARAMS
# =========================
DATA_DIR = r"/kaggle/input/mangoleafdisease-classes/MangoLeafBD Dataset/MangoLeafBD Dataset"
# change the folder name to whatever shows in /kaggle/input

IMG_SIZE = (224, 224)     # keep 224 for AlexNet/ViT/DaViT fairness
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

# =========================
# 2) READ FILE PATHS + LABELS
# =========================
class_names = sorted([
    d for d in os.listdir(DATA_DIR)
    if os.path.isdir(os.path.join(DATA_DIR, d))
])
num_classes = len(class_names)
print("Classes:", class_names)

label_to_index = {c: i for i, c in enumerate(class_names)}

filepaths, labels = [], []
exts = (".jpg", ".jpeg", ".png", ".bmp", ".webp")

for c in class_names:
    cdir = os.path.join(DATA_DIR, c)
    for f in os.listdir(cdir):
        if f.lower().endswith(exts):
            filepaths.append(os.path.join(cdir, f))
            labels.append(label_to_index[c])

filepaths = np.array(filepaths)
labels = np.array(labels)

print("Total images:", len(filepaths))

# =========================
# 3) STRATIFIED SPLIT
# =========================
from sklearn.model_selection import train_test_split

train_files, temp_files, train_labels, temp_labels = train_test_split(
    filepaths, labels, test_size=0.30, stratify=labels, random_state=SEED
)
val_files, test_files, val_labels, test_labels = train_test_split(
    temp_files, temp_labels, test_size=0.50, stratify=temp_labels, random_state=SEED
)

print("Train:", len(train_files))
print("Val:", len(val_files))
print("Test:", len(test_files))


# =========================
# 4) DATASET LOADER
# =========================
def load_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img.set_shape([None, None, 3])
    img = tf.image.resize(img, IMG_SIZE)
    img = tf.cast(img, tf.float32)

    # EfficientNetV2 preprocessing
    img = keras.applications.efficientnet_v2.preprocess_input(img)

    label = tf.one_hot(label, num_classes)
    return img, label

def make_ds(files, labs, training=False):
    ds = tf.data.Dataset.from_tensor_slices((files, labs))
    if training:
        ds = ds.shuffle(len(files), seed=SEED)
    ds = ds.map(load_image, num_parallel_calls=AUTOTUNE)
    ds = ds.batch(BATCH_SIZE, drop_remainder=training).prefetch(AUTOTUNE)
    return ds

train_ds = make_ds(train_files, train_labels, training=True)
val_ds   = make_ds(val_files, val_labels, training=False)
test_ds  = make_ds(test_files, test_labels, training=False)


# =========================
# 5) AUGMENTATION (effective, not too heavy)
# =========================
augment = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.10),
    layers.RandomZoom(0.12),
    layers.RandomContrast(0.08),
    layers.RandomTranslation(0.05, 0.05),
], name="augmentation")


# =========================
# 7) BUILD ALEXNET MODEL (stable head)
# =========================
inputs = keras.Input(shape=(*IMG_SIZE, 3))

# augmentation first (same as your pipeline)
x = augment(inputs, training=True)

# ---- AlexNet feature extractor ----
# Block 1
x = layers.Conv2D(96, kernel_size=11, strides=4, padding="valid", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D(pool_size=3, strides=2)(x)

# Block 2
x = layers.Conv2D(256, kernel_size=5, strides=1, padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D(pool_size=3, strides=2)(x)

# Block 3
x = layers.Conv2D(384, kernel_size=3, padding="same", activation="relu")(x)
x = layers.Conv2D(384, kernel_size=3, padding="same", activation="relu")(x)
x = layers.Conv2D(256, kernel_size=3, padding="same", activation="relu")(x)
x = layers.MaxPooling2D(pool_size=3, strides=2)(x)

# ---- Classifier head ----
x = layers.Flatten()(x)
x = layers.Dense(4096, activation="relu")(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(4096, activation="relu")(x)
x = layers.Dropout(0.5)(x)

outputs = layers.Dense(num_classes, activation="softmax")(x)

model = keras.Model(inputs, outputs, name="AlexNet")
model.summary()

# =========================
# 8) CLASS WEIGHTS (helps if imbalance)
# =========================
from sklearn.utils.class_weight import compute_class_weight
cw = compute_class_weight("balanced", classes=np.arange(num_classes), y=train_labels)
class_weights = {i: w for i, w in enumerate(cw)}
print("Class weights:", class_weights)


# =========================
# 9) STAGE 1 TRAIN (freeze backbone)
# =========================
loss_stage1 = keras.losses.CategoricalCrossentropy(label_smoothing=0.02)

model.compile(
    optimizer=keras.optimizers.AdamW(learning_rate=3e-4, weight_decay=1e-4),
    loss=loss_stage1,
    metrics=["accuracy"]
)

cbs1 = [
    keras.callbacks.ModelCheckpoint("best_stage1.keras", save_best_only=True,
                                    monitor="val_accuracy", mode="max"),
    keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5,
                                      patience=2, verbose=1),
    keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=5,
                                  restore_best_weights=True),
]

history1 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=12,
    callbacks=cbs1,
    class_weight=class_weights
)

# =========================
# 10) STAGE 2 FINE-TUNE (unfreeze more)
# =========================
model.trainable = True
for layer in model.layers[:-10]:
    layer.trainable = False

model.compile(
    optimizer=keras.optimizers.AdamW(learning_rate=5e-5, weight_decay=5e-6),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

cbs2 = [
    keras.callbacks.ModelCheckpoint("best_finetune.keras", save_best_only=True,
                                    monitor="val_accuracy", mode="max"),
    keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5,
                                      patience=2, verbose=1),
    keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=7,
                                  restore_best_weights=True),
]

history2 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=25,
    callbacks=cbs2,
    class_weight=class_weights
)

# =========================
# 11) NORMAL TEST ACC
# =========================
model = keras.models.load_model("best_finetune.keras", compile=False)
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

test_loss, test_acc = model.evaluate(test_ds, verbose=1)
print("TEST ACC (normal):", test_acc)

# =========================
# 12) TEST-TIME AUGMENTATION (TTA) ACC  [FIXED for TF 2.18 / Keras 3]
#    (no keras layers inside ds.map)
# =========================
tta_aug = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.06),
    layers.RandomZoom(0.06),
], name="tta_aug")

def tta_predict(model, ds, runs=8):
    preds_sum = None
    for r in range(runs):
        preds_run = []
        for x_batch, _ in ds:
            x_aug = tta_aug(x_batch, training=True)   # augment in eager mode
            preds = model.predict(x_aug, verbose=0)
            preds_run.append(preds)
        preds_run = np.concatenate(preds_run, axis=0)

        if preds_sum is None:
            preds_sum = preds_run
        else:
            preds_sum += preds_run

    return preds_sum / runs

tta_preds = tta_predict(model, test_ds, runs=8)
tta_pred_labels = np.argmax(tta_preds, axis=1)

# true labels (for one-hot categorical labels)
true_labels = []
for _, y in test_ds:
    true_labels.extend(np.argmax(y.numpy(), axis=1))
true_labels = np.array(true_labels)

tta_acc = np.mean(true_labels == tta_pred_labels)
print("TEST ACC (TTA):", tta_acc)


import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report

# =========================
# 1) ACCURACY + LOSS GRAPHS
# =========================
def plot_history(histories):
    acc, val_acc, loss, val_loss = [], [], [], []

    for h in histories:
        acc      += h.history.get("accuracy", [])
        val_acc  += h.history.get("val_accuracy", [])
        loss     += h.history.get("loss", [])
        val_loss += h.history.get("val_loss", [])

    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 8))
    plt.rcParams.update({"font.size": 16})
    plt.plot(epochs, acc, label="Train Accuracy", linewidth=2)
    plt.plot(epochs, val_acc, label="Val Accuracy", linewidth=2)
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.title("Training vs Validation Accuracy")
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(12, 8))
    plt.rcParams.update({"font.size": 16})
    plt.plot(epochs, loss, label="Train Loss", linewidth=2)
    plt.plot(epochs, val_loss, label="Val Loss", linewidth=2)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title("Training vs Validation Loss")
    plt.legend()
    plt.grid(True)
    plt.show()


# If you trained in 2 stages:
plot_history([history1, history2])

# If you trained only once, use:
# plot_history([history])

# =========================
# 2) CONFUSION MATRIX + REPORT
#    (works for BOTH one-hot and sparse labels)
# =========================
y_true = []
y_pred = []

for x_batch, y_batch in test_ds:
    preds = model.predict(x_batch, verbose=0)

    # handle label format automatically
    if len(y_batch.shape) > 1 and y_batch.shape[-1] > 1:
        # one-hot labels
        y_true.extend(np.argmax(y_batch.numpy(), axis=1))
    else:
        # sparse integer labels
        y_true.extend(y_batch.numpy().astype(int))

    y_pred.extend(np.argmax(preds, axis=1))

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(12, 8))
plt.rcParams.update({"font.size": 16})
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix")
plt.colorbar()

tick_marks = np.arange(num_classes)
plt.xticks(tick_marks, class_names, rotation=45, ha="right")
plt.yticks(tick_marks, class_names)

# write numbers in cells
thresh = cm.max() / 2.0
for i in range(num_classes):
    for j in range(num_classes):
        plt.text(
            j, i, cm[i, j],
            ha="center", va="center",
            color="white" if cm[i, j] > thresh else "black"
        )

plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.tight_layout()
plt.show()


import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# =========================
# ROC EVALUATION (Multiclass OvR) - ROBUST VERSION
# Works for BOTH one-hot and sparse labels
# =========================

y_true_list = []
y_score_list = []

for x_batch, y_batch in test_ds:
    # predictions
    y_score_list.append(model.predict(x_batch, verbose=0))

    # handle label format
    yb = y_batch.numpy()
    if len(yb.shape) > 1 and yb.shape[-1] > 1:
        # one-hot
        y_true_list.append(yb)
    else:
        # sparse -> binarize
        y_true_list.append(
            label_binarize(yb.astype(int).reshape(-1), classes=range(num_classes))
        )

y_true  = np.concatenate(y_true_list, axis=0)    # (N, C)
y_score = np.concatenate(y_score_list, axis=0)   # (N, C)

# =========================
# Compute ROC for each class
# =========================
fpr, tpr, roc_auc = {}, {}, {}

for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Micro-average ROC
fpr["micro"], tpr["micro"], _ = roc_curve(y_true.ravel(), y_score.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Macro-average ROC
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(num_classes)]))
mean_tpr = np.zeros_like(all_fpr)

for i in range(num_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

mean_tpr /= num_classes
fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

# =========================
# PLOT ROC
# =========================
plt.figure(figsize=(12, 8))
plt.rcParams.update({"font.size": 19})

for i, cls in enumerate(class_names):
    plt.plot(
        fpr[i], tpr[i],
        linewidth=2,
        label=f"{cls} (AUC = {roc_auc[i]:.3f})"
    )

plt.plot(
    fpr["micro"], tpr["micro"],
    linestyle="--", linewidth=3,
    label=f"Micro-average (AUC = {roc_auc['micro']:.3f})"
)
plt.plot(
    fpr["macro"], tpr["macro"],
    linestyle="--", linewidth=3,
    label=f"Macro-average (AUC = {roc_auc['macro']:.3f})"
)

plt.plot([0, 1], [0, 1], linestyle=":", linewidth=2)

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Multiclass ROC Curve (One-vs-Rest)")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2

# =========================
# XAI: GRAD-CAM FOR AlexNet
# =========================

# Find last Conv2D layer name automatically
def get_last_conv_layer(model):
    for layer in reversed(model.layers):
        if isinstance(layer, tf.keras.layers.Conv2D):
            return layer.name
    raise ValueError("No Conv2D layer found in model.")

LAST_CONV_NAME = get_last_conv_layer(model)
print("Grad-CAM using last conv layer:", LAST_CONV_NAME)


def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = tf.keras.models.Model(
        [model.inputs],
        [model.get_layer(last_conv_layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(predictions[0])
        class_score = predictions[:, pred_index]

    grads = tape.gradient(class_score, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    heatmap = tf.maximum(heatmap, 0) / (tf.reduce_max(heatmap) + 1e-8)
    return heatmap.numpy()


def show_gradcam_on_image(img, heatmap, alpha=0.45):
    # resize heatmap
    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))

    # colorize
    heatmap_color = cv2.applyColorMap(np.uint8(255 * heatmap), cv2.COLORMAP_JET)
    heatmap_color = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)

    # overlay
    overlay = heatmap_color * alpha + img
    overlay = np.clip(overlay / 255.0, 0, 1)

    return overlay


# =========================
# VISUALIZE ON SAMPLE TEST IMAGES
# =========================
plt.figure(figsize=(12, 10))

for x_batch, y_batch in test_ds.take(1):
    preds = model.predict(x_batch, verbose=0)

    pred_labels = np.argmax(preds, axis=1)
    true_labels = np.argmax(y_batch.numpy(), axis=1)

    for i in range(min(6, x_batch.shape[0])):
        img_proc = x_batch[i].numpy()

        # -------- Convert back for display --------
        # you used EfficientNetV2 preprocess_input -> values approx in [-1,1]
        # so bring back to [0,255] for visualization
        img_disp = (img_proc + 1.0) * 127.5
        img_disp = np.clip(img_disp, 0, 255).astype(np.uint8)

        heatmap = make_gradcam_heatmap(
            np.expand_dims(x_batch[i], axis=0),
            model,
            LAST_CONV_NAME
        )

        overlay = show_gradcam_on_image(img_disp, heatmap)

        plt.subplot(2, 3, i+1)
        plt.imshow(overlay)
        plt.axis("off")
        plt.title(
            f"True: {class_names[true_labels[i]]}\nPred: {class_names[pred_labels[i]]}"
        )

plt.tight_layout()
plt.show()
```

```python
# =========================
# 1) IMPORT LIBRARIES
# =========================
import os, random
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# =========================
# 2) SET DATA PATH + READ FILES
# =========================
DATA_DIR = r"/kaggle/input/mangoleafdisease-classes/MangoLeafBD Dataset/MangoLeafBD Dataset"
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

class_names = sorted([
    d for d in os.listdir(DATA_DIR)
    if os.path.isdir(os.path.join(DATA_DIR, d))
])
num_classes = len(class_names)
print("Classes:", class_names)

label_to_index = {c:i for i,c in enumerate(class_names)}

filepaths, labels = [], []
exts = (".jpg", ".jpeg", ".png", ".bmp", ".webp")

for c in class_names:
    cdir = os.path.join(DATA_DIR, c)
    for f in os.listdir(cdir):
        if f.lower().endswith(exts):
            filepaths.append(os.path.join(cdir, f))
            labels.append(label_to_index[c])

filepaths = np.array(filepaths)
labels = np.array(labels)
print("Total images:", len(filepaths))

# =========================
# 3) STRATIFIED SPLIT
# =========================
train_files, temp_files, train_labels, temp_labels = train_test_split(
    filepaths, labels,
    test_size=0.30,
    stratify=labels,
    random_state=SEED
)

val_files, test_files, val_labels, test_labels = train_test_split(
    temp_files, temp_labels,
    test_size=0.50,
    stratify=temp_labels,
    random_state=SEED
)

print("Train:", len(train_files))
print("Val  :", len(val_files))
print("Test :", len(test_files))

# =========================
# 4) PREPROCESSING + tf.data (RAW IMAGES)
# =========================
def load_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img.set_shape([None, None, 3])
    img = tf.image.resize(img, IMG_SIZE)
    img = tf.cast(img, tf.float32)  # raw [0..255]
    label = tf.one_hot(label, num_classes)
    return img, label

def make_ds(files, labs, training=False):
    ds = tf.data.Dataset.from_tensor_slices((files, labs))
    if training:
        ds = ds.shuffle(len(files), seed=SEED)
    ds = ds.map(load_image, num_parallel_calls=AUTOTUNE)
    ds = ds.batch(BATCH_SIZE, drop_remainder=training).prefetch(AUTOTUNE)
    return ds

train_ds = make_ds(train_files, train_labels, training=True)
val_ds   = make_ds(val_files, val_labels, training=False)
test_ds  = make_ds(test_files, test_labels, training=False)

# class weights (small data)
cw = compute_class_weight("balanced", classes=np.arange(num_classes), y=train_labels)
class_weights = {i: w for i, w in enumerate(cw)}
print("Class weights:", class_weights)

# =========================
# 5) STRONG BUT SAFE AUGMENTATION (MODEL SIDE)
# =========================
augment = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.12),
    layers.RandomZoom(0.15),
    layers.RandomContrast(0.12),
    layers.RandomTranslation(0.06, 0.06),
], name="augmentation")

# =========================
# 6) BUILD ANTI-OVERFIT VGG16
# =========================
base = keras.applications.VGG16(
    include_top=False,
    weights="imagenet",
    input_shape=(*IMG_SIZE, 3)
)
base.trainable = False

reg = keras.regularizers.l2(2e-4)  # stronger L2

inputs = keras.Input(shape=(*IMG_SIZE, 3))
x = augment(inputs)

# preprocess AFTER augmentation
x = keras.applications.vgg16.preprocess_input(x)

x = base(x, training=False)

x = layers.GlobalAveragePooling2D()(x)

# smaller + more regularized head
x = layers.Dropout(0.55)(x)
x = layers.Dense(256, activation="relu", kernel_regularizer=reg)(x)
x = layers.Dropout(0.40)(x)

outputs = layers.Dense(num_classes, activation="softmax")(x)

model = keras.Model(inputs, outputs, name="VGG16_anti_overfit")
model.summary()

# =========================
# 7) CALLBACKS
# =========================
def get_callbacks(save_name):
    return [
        keras.callbacks.ModelCheckpoint(
            save_name, monitor="val_accuracy",
            save_best_only=True, mode="max", verbose=1
        ),
        keras.callbacks.ReduceLROnPlateau(
            monitor="val_loss", factor=0.5,
            patience=2, verbose=1
        ),
        keras.callbacks.EarlyStopping(
            monitor="val_loss", patience=5,
            restore_best_weights=True, verbose=1
        ),
    ]

# =========================
# 8) STAGE 1 TRAIN (FROZEN)
# =========================
loss_fn = keras.losses.CategoricalCrossentropy(label_smoothing=0.03)

model.compile(
    optimizer=keras.optimizers.AdamW(2e-4, weight_decay=1e-4),
    loss=loss_fn,
    metrics=["accuracy"]
)

history1 = model.fit(
    train_ds,                     # ✅ MixUp removed
    validation_data=val_ds,
    epochs=12,
    callbacks=get_callbacks("vgg16_stage1.keras"),
    class_weight=class_weights
)

# =========================
# 9) STAGE 2 FINE-TUNE (ONLY LAST CONV LAYERS)
# =========================
base.trainable = True
for layer in base.layers:
    if "block5_conv3" not in layer.name and "block5_conv2" not in layer.name:
        layer.trainable = False  # fine-tune only top 2 conv layers

model.compile(
    optimizer=keras.optimizers.AdamW(2e-5, weight_decay=5e-6),
    loss=loss_fn,
    metrics=["accuracy"]
)

history2 = model.fit(
    train_ds,                     # ✅ MixUp removed
    validation_data=val_ds,
    epochs=20,
    callbacks=get_callbacks("vgg16_best.keras"),
    class_weight=class_weights
)

# =========================
# 10) TEST
# =========================
model = keras.models.load_model("vgg16_best.keras", compile=False)
model.compile(optimizer="adam", loss=loss_fn, metrics=["accuracy"])
test_loss, test_acc = model.evaluate(test_ds, verbose=1)
print("VGG16 TEST ACC (no MixUp):", test_acc)
```

```python
# ==========================================================
# EVALUATION + GRAPHS + CONFUSION MATRIX + ROC (font=19)
# ==========================================================
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from sklearn.preprocessing import label_binarize

# --------------------------
# 1) Reload best model + evaluate
# --------------------------
model = keras.models.load_model("vgg16_best.keras", compile=False)
model.compile(optimizer="adam", loss=loss_fn, metrics=["accuracy"])

test_loss, test_acc = model.evaluate(test_ds, verbose=1)
print("✅ TEST ACC:", test_acc)
print("✅ TEST LOSS:", test_loss)


# --------------------------
# 2) ACCURACY + LOSS GRAPHS
# --------------------------
def plot_history(histories):
    acc, val_acc, loss, val_loss = [], [], [], []

    for h in histories:
        acc      += h.history.get("accuracy", [])
        val_acc  += h.history.get("val_accuracy", [])
        loss     += h.history.get("loss", [])
        val_loss += h.history.get("val_loss", [])

    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 8))
    plt.rcParams.update({"font.size": 19})
    plt.plot(epochs, acc, label="Train Accuracy", linewidth=2)
    plt.plot(epochs, val_acc, label="Val Accuracy", linewidth=2)
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.title("Training vs Validation Accuracy")
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(12, 8))
    plt.rcParams.update({"font.size": 19})
    plt.plot(epochs, loss, label="Train Loss", linewidth=2)
    plt.plot(epochs, val_loss, label="Val Loss", linewidth=2)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title("Training vs Validation Loss")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_history([history1, history2])


# --------------------------
# 3) CONFUSION MATRIX + REPORT
# --------------------------
y_true, y_pred = [], []

for x_batch, y_batch in test_ds:
    preds = model.predict(x_batch, verbose=0)
    y_true.extend(np.argmax(y_batch.numpy(), axis=1))
    y_pred.extend(np.argmax(preds, axis=1))

print("\n✅ Classification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(12, 8))
plt.rcParams.update({"font.size": 19})
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix")
plt.colorbar()

tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=45, ha="right")
plt.yticks(tick_marks, class_names)

thresh = cm.max() / 2.0
for i in range(len(class_names)):
    for j in range(len(class_names)):
        plt.text(
            j, i, cm[i, j],
            ha="center", va="center",
            color="white" if cm[i, j] > thresh else "black"
        )

plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.tight_layout()
plt.show()


# --------------------------
# 4) ROC CURVE (Multiclass OvR)
# --------------------------
y_true_list, y_score_list = [], []

for x_batch, y_batch in test_ds:
    y_true_list.append(y_batch.numpy())
    y_score_list.append(model.predict(x_batch, verbose=0))

y_true  = np.concatenate(y_true_list, axis=0)   # (N, C)
y_score = np.concatenate(y_score_list, axis=0) # (N, C)

fpr, tpr, roc_auc = {}, {}, {}
num_classes = len(class_names)

for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# micro-average
fpr["micro"], tpr["micro"], _ = roc_curve(y_true.ravel(), y_score.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# macro-average
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(num_classes)]))
mean_tpr = np.zeros_like(all_fpr)

for i in range(num_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

mean_tpr /= num_classes
fpr["macro"], tpr["macro"] = all_fpr, mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])


plt.figure(figsize=(12, 8))
plt.rcParams.update({"font.size": 19})

for i, cls in enumerate(class_names):
    plt.plot(
        fpr[i], tpr[i], linewidth=2,
        label=f"{cls} (AUC = {roc_auc[i]:.3f})"
    )

plt.plot(
    fpr["micro"], tpr["micro"], linestyle="--", linewidth=3,
    label=f"Micro avg (AUC = {roc_auc['micro']:.3f})"
)
plt.plot(
    fpr["macro"], tpr["macro"], linestyle="--", linewidth=3,
    label=f"Macro avg (AUC = {roc_auc['macro']:.3f})"
)

plt.plot([0, 1], [0, 1], linestyle=":", linewidth=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Multiclass ROC Curve (One-vs-Rest)")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
# ====== Libraries needed for DenseNet201 (TensorFlow + Kaggle) ======

import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight

SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)

# =========================
# 1) PATHS / PARAMS
# =========================
DATA_DIR = r"/kaggle/input/mangoleafdisease-classes/MangoLeafBD Dataset/MangoLeafBD Dataset"
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

# =========================
# 2) READ FILE PATHS + LABELS
# =========================
class_names = sorted([
    d for d in os.listdir(DATA_DIR)
    if os.path.isdir(os.path.join(DATA_DIR, d))
])
num_classes = len(class_names)
print("Classes:", class_names)

label_to_index = {c: i for i, c in enumerate(class_names)}

filepaths, labels = [], []
exts = (".jpg", ".jpeg", ".png", ".bmp", ".webp")

for c in class_names:
    cdir = os.path.join(DATA_DIR, c)
    for f in os.listdir(cdir):
        if f.lower().endswith(exts):
            filepaths.append(os.path.join(cdir, f))
            labels.append(label_to_index[c])

filepaths = np.array(filepaths)
labels = np.array(labels)

print("Total images:", len(filepaths))

# =========================
# 3) STRATIFIED SPLIT (70/15/15)
# =========================
train_files, temp_files, train_labels, temp_labels = train_test_split(
    filepaths, labels, test_size=0.30, stratify=labels, random_state=SEED
)
val_files, test_files, val_labels, test_labels = train_test_split(
    temp_files, temp_labels, test_size=0.50, stratify=temp_labels, random_state=SEED
)

print("Train:", len(train_files))
print("Val:", len(val_files))
print("Test:", len(test_files))

# =========================
# 4) DATASET LOADER (DenseNet preprocessing)
# =========================
def load_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img.set_shape([None, None, 3])
    img = tf.image.resize(img, IMG_SIZE)
    img = tf.cast(img, tf.float32)

    img = keras.applications.densenet.preprocess_input(img)

    label = tf.one_hot(label, num_classes)
    return img, label

def make_ds(files, labs, training=False):
    ds = tf.data.Dataset.from_tensor_slices((files, labs))
    if training:
        ds = ds.shuffle(len(files), seed=SEED)
    ds = ds.map(load_image, num_parallel_calls=AUTOTUNE)
    ds = ds.batch(BATCH_SIZE, drop_remainder=training).prefetch(AUTOTUNE)
    return ds

train_ds = make_ds(train_files, train_labels, training=True)
val_ds   = make_ds(val_files, val_labels, training=False)
test_ds  = make_ds(test_files, test_labels, training=False)

# =========================
# 5) AUGMENTATION (ONLY TRAIN)
# =========================
augment = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.12),
    layers.RandomZoom(0.15),
    layers.RandomContrast(0.10),
    layers.RandomTranslation(0.06, 0.06),
], name="augmentation")

# =========================
# 6) MixUp (reduces overfit a lot)
# =========================
def mixup_batch(x, y, alpha=0.2):
    """Applies MixUp to a batch."""
    batch_size = tf.shape(x)[0]
    lam = tfp.distributions.Beta(alpha, alpha).sample([batch_size])
    lam_x = tf.reshape(lam, (batch_size, 1, 1, 1))
    lam_y = tf.reshape(lam, (batch_size, 1))

    idx = tf.random.shuffle(tf.range(batch_size))
    x2 = tf.gather(x, idx)
    y2 = tf.gather(y, idx)

    x_mix = lam_x * x + (1 - lam_x) * x2
    y_mix = lam_y * y + (1 - lam_y) * y2
    return x_mix, y_mix

# We need tensorflow_probability for MixUp Beta distribution
!pip -q install tensorflow_probability
import tensorflow_probability as tfp

def apply_mixup(ds):
    return ds.map(lambda x, y: mixup_batch(x, y), num_parallel_calls=AUTOTUNE)

train_ds_mix = apply_mixup(train_ds)

# =========================
# 7) BUILD DenseNet201 MODEL (better head)
# =========================
base = keras.applications.DenseNet201(
    include_top=False,
    weights="imagenet",
    input_shape=(*IMG_SIZE, 3)
)
base.trainable = False

inputs = keras.Input(shape=(*IMG_SIZE, 3))
x = augment(inputs)                 # train-time augment only
x = base(x, training=False)

x = layers.BatchNormalization()(x)

# SE block
se = layers.GlobalAveragePooling2D()(x)
se = layers.Dense(se.shape[-1] // 8, activation="relu")(se)
se = layers.Dense(x.shape[-1], activation="sigmoid")(se)
se = layers.Reshape((1, 1, x.shape[-1]))(se)
x = layers.Multiply()([x, se])

x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.45)(x)

x = layers.Dense(512, activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.35)(x)

x = layers.Dense(256, activation="relu")(x)
x = layers.Dropout(0.30)(x)

outputs = layers.Dense(num_classes, activation="softmax")(x)
model = keras.Model(inputs, outputs, name="DenseNet201")
model.summary()

# =========================
# 8) CLASS WEIGHTS
# =========================
cw = compute_class_weight("balanced", classes=np.arange(num_classes), y=train_labels)
class_weights = {i: w for i, w in enumerate(cw)}
print("Class weights:", class_weights)

# =========================
# 9) Stage 1 TRAIN (freeze backbone, cosine LR, strong reg)
# =========================
epochs_stage1 = 12
steps_per_epoch = int(np.ceil(len(train_files) / BATCH_SIZE))

cosine_lr = keras.optimizers.schedules.CosineDecay(
    initial_learning_rate=1e-4,
    decay_steps=epochs_stage1 * steps_per_epoch
)

loss_stage1 = keras.losses.CategoricalCrossentropy(label_smoothing=0.05)

optimizer1 = keras.optimizers.AdamW(
    learning_rate=cosine_lr,
    weight_decay=5e-4,
    use_ema=True,  # EMA helps generalization
    ema_momentum=0.999
)

model.compile(
    optimizer=optimizer1,
    loss=loss_stage1,
    metrics=["accuracy"]
)

cbs1 = [
    keras.callbacks.ModelCheckpoint("best_stage1.keras", save_best_only=True,
                                    monitor="val_accuracy", mode="max"),
    keras.callbacks.EarlyStopping(monitor="val_loss", patience=4,
                                  restore_best_weights=True),
]

history1 = model.fit(
    train_ds_mix,   # use MixUp dataset
    validation_data=val_ds,
    epochs=epochs_stage1,
    callbacks=cbs1,
    class_weight=class_weights
)

# =========================
# 10) Stage 2 Fine-tune (unfreeze last 80 layers slowly)
# =========================
base.trainable = True
for layer in base.layers[:-80]:
    layer.trainable = False

epochs_stage2 = 25
cosine_lr2 = keras.optimizers.schedules.CosineDecay(
    initial_learning_rate=3e-5,
    decay_steps=epochs_stage2 * steps_per_epoch
)

optimizer2 = keras.optimizers.AdamW(
    learning_rate=cosine_lr2,
    weight_decay=1e-4,
    use_ema=True,
    ema_momentum=0.999
)

model.compile(
    optimizer=optimizer2,
    loss=keras.losses.CategoricalCrossentropy(label_smoothing=0.03),
    metrics=["accuracy"]
)

cbs2 = [
    keras.callbacks.ModelCheckpoint("best_finetune.keras", save_best_only=True,
                                    monitor="val_accuracy", mode="max"),
    keras.callbacks.EarlyStopping(monitor="val_loss", patience=6,
                                  restore_best_weights=True),
]

history2 = model.fit(
    train_ds_mix,
    validation_data=val_ds,
    epochs=epochs_stage2,
    callbacks=cbs2,
    class_weight=class_weights
)

# =========================
# 11) TEST ACC
# =========================
model = keras.models.load_model("best_finetune.keras", compile=False)
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

test_loss, test_acc = model.evaluate(test_ds, verbose=1)
print("TEST ACC:", test_acc)

# =========================
# 12) CONFUSION MATRIX + REPORT
# =========================
y_true, y_pred = [], []

for x_batch, y_batch in test_ds:
    preds = model.predict(x_batch, verbose=0)
    y_true.extend(np.argmax(y_batch.numpy(), axis=1))
    y_pred.extend(np.argmax(preds, axis=1))

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)
print("\nConfusion Matrix:\n", cm)
```

```python
# =========================
# ONE CELL: Curves + Confusion Matrix + ROC
# fig size = (12,8), font size = 19
# =========================
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, roc_auc_score

# global plot style
plt.rcParams.update({"font.size": 19})

# ---------- 1) ACC & LOSS CURVES (Stage1 + Stage2) ----------
def plot_history_two_stages(h1, h2, title="DenseNet201 Training Curves"):
    acc      = h1.history["accuracy"] + h2.history["accuracy"]
    val_acc  = h1.history["val_accuracy"] + h2.history["val_accuracy"]
    loss     = h1.history["loss"] + h2.history["loss"]
    val_loss = h1.history["val_loss"] + h2.history["val_loss"]

    epochs = range(1, len(acc) + 1)
    split_ep = len(h1.history["accuracy"])

    # Accuracy
    plt.figure(figsize=(12,8))
    plt.plot(epochs, acc, label="Train Accuracy", linewidth=2)
    plt.plot(epochs, val_acc, label="Val Accuracy", linewidth=2)
    plt.axvline(split_ep, color="gray", linestyle="--", linewidth=2, label="Stage-2 start")
    plt.title(title + " (Accuracy)")
    plt.xlabel("Epochs"); plt.ylabel("Accuracy")
    plt.legend(); plt.grid(True); plt.show()

    # Loss
    plt.figure(figsize=(12,8))
    plt.plot(epochs, loss, label="Train Loss", linewidth=2)
    plt.plot(epochs, val_loss, label="Val Loss", linewidth=2)
    plt.axvline(split_ep, color="gray", linestyle="--", linewidth=2, label="Stage-2 start")
    plt.title(title + " (Loss)")
    plt.xlabel("Epochs"); plt.ylabel("Loss")
    plt.legend(); plt.grid(True); plt.show()

plot_history_two_stages(history1, history2)


# ---------- 2) CONFUSION MATRIX + REPORT ----------
y_true, y_pred = [], []

for x_batch, y_batch in test_ds:
    preds = model.predict(x_batch, verbose=0)
    y_true.extend(np.argmax(y_batch.numpy(), axis=1))
    y_pred.extend(np.argmax(preds, axis=1))

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(12,8))
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix")
plt.colorbar()

tick_marks = np.arange(num_classes)
plt.xticks(tick_marks, class_names, rotation=45, ha="right")
plt.yticks(tick_marks, class_names)

thresh = cm.max() / 2
for i in range(num_classes):
    for j in range(num_classes):
        plt.text(
            j, i, cm[i, j],
            ha="center", va="center",
            color="white" if cm[i, j] > thresh else "black"
        )

plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.tight_layout()
plt.show()


# ---------- 3) MULTICLASS ROC (OvR) + MACRO/MICRO AUC ----------
y_true_list, y_score_list = [], []
for x_batch, y_batch in test_ds:
    y_score_list.append(model.predict(x_batch, verbose=0))
    y_true_list.append(y_batch.numpy())   # one-hot

y_true  = np.concatenate(y_true_list, axis=0)   # (N, C)
y_score = np.concatenate(y_score_list, axis=0) # (N, C)

fpr, tpr, roc_auc = {}, {}, {}
for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# micro-average
fpr["micro"], tpr["micro"], _ = roc_curve(y_true.ravel(), y_score.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# macro-average
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(num_classes)]))
mean_tpr = np.zeros_like(all_fpr)
for i in range(num_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
mean_tpr /= num_classes
fpr["macro"], tpr["macro"] = all_fpr, mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

plt.figure(figsize=(12,8))
for i, cls in enumerate(class_names):
    plt.plot(fpr[i], tpr[i], linewidth=2, label=f"{cls} (AUC={roc_auc[i]:.3f})")

plt.plot(fpr["micro"], tpr["micro"], linestyle="--", linewidth=3,
         label=f"Micro-average (AUC={roc_auc['micro']:.3f})")
plt.plot(fpr["macro"], tpr["macro"], linestyle="--", linewidth=3,
         label=f"Macro-average (AUC={roc_auc['macro']:.3f})")

plt.plot([0,1],[0,1], linestyle=":", linewidth=2)
plt.xlim([0.0,1.0]); plt.ylim([0.0,1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Multiclass ROC Curve (OvR)")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

macro_auc = roc_auc_score(y_true, y_score, average="macro", multi_class="ovr")
micro_auc = roc_auc_score(y_true, y_score, average="micro", multi_class="ovr")
print("Macro AUC:", macro_auc)
print("Micro AUC:", micro_auc)
```
